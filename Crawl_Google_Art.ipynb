# Googles Art & Culture

### All artworks were crawled from the webpage: https://artsandculture.google.com/partner



<img src="screenshot.png" style="width: 900px;"/>

```python
import requests 
import lxml.html
import urllib
import os
import re
from urllib import request
```

```python
google_url = "https://artsandculture.google.com/partner"
```

```python
def read_html(url):
    source_code = requests.get(url) 
    html_elements = lxml.html.fromstring(source_code.content) 
    return html_elements
```

```python
# method to create folder for every gallery

def create_folder(directory):
    try:
        if not os.path.exists(directory): 
            os.makedirs(directory) 
    except OSError:
        print ('Error: Creating directory. ' +  directory)
```

# Start Crawling from here

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from selenium.webdriver.common.by import By


driver = webdriver.Chrome(executable_path=r"C:\Program Files\Driver\chromedriver.exe")
driver.get('https://artsandculture.google.com/partner')
driver.maximize_window()


SCROLL_PAUSE_TIME = 20

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    #Find all links to galleries
    all_links = driver.find_elements_by_xpath("//a[@class='e0WtYb'][@href]")
    
    #Select link in all links 
    for links in all_links:

        time.sleep(SCROLL_PAUSE_TIME)
        link_to_gallery = links.get_attribute('href')
        print(link_to_gallery)

        read_gallery_html = read_html(link_to_gallery) 
        xpath_alle_ansehen = "//a[@class='e0WtYb qPaDxc']/@href"
        link_alle_ansehen = "https://artsandculture.google.com" + "".join(read_gallery_html.xpath(xpath_alle_ansehen))
        print(link_alle_ansehen)

        #Name der Gallerie extrahieren
        read_html_alle_ansehen = read_html(link_alle_ansehen)     
        xpath_name_of_gallery = read_html_alle_ansehen.xpath("//h1/text()")
        gallerie_name, zusatzinfo = " ".join(xpath_name_of_gallery).split('|') 
        gallerie_name_cleaned = re.sub(r'\W+', '_', gallerie_name)[:-1]
        print(gallerie_name_cleaned)

        #Create folder: Name and Location of Gallerie     
        create_folder('./Google_Art/' + gallerie_name_cleaned)

        folder = './Google_Art/' + gallerie_name_cleaned

        #Select first element on "alle ansehen Webseite" where all artworks are stored
        xpath_to_first_element = "//*[@id='yDmH0d']/div[3]/div/div/div[2]/div/ul/li[1]/a/@href"
        first_element_in_view_all = "https://artsandculture.google.com" + "".join(read_html_alle_ansehen.xpath(xpath_to_first_element))
        print(first_element_in_view_all)

        images_website = webdriver.Chrome(executable_path=r"C:\Program Files\Driver\chromedriver.exe")
        images_website.get(first_element_in_view_all)
        images_website.maximize_window()

        # Get scroll height
        last_height_image_website = images_website.execute_script("return document.body.scrollHeight")
        while True:
            all_links_image_website = images_website.find_elements_by_xpath("//a[@class='e0WtYb DRVwp PJLMUc'][@href]") 

            #Select link in all links 
            for image in all_links_image_website:
                time.sleep(SCROLL_PAUSE_TIME)
                link_to_image = image.get_attribute('href')
                print(link_to_image)

                # URL to artwork
                read_html_to_artwork = read_html(link_to_image) 
                xpath_to_image = "//img/@src"

                url_to_image2 = "".join(read_html_to_artwork.xpath(xpath_to_image))
                url_to_image_cleaned = url_to_image2.replace('//', '')
                url_to_image_cleaned = "http://" + url_to_image_cleaned 

                # Name of painting  
                xpath_to_artwork_name = "/html/body/div[3]/div[3]/div/div/div[6]/div/header/div[2]/div[1]/div[1]/h1/text()"
                url_to_artwork_name = "".join(read_html_to_artwork.xpath(xpath_to_artwork_name))
                name_artwork_cleaned = re.sub(r'\W+', '_', url_to_artwork_name)
                #print(name_artwork_cleaned)            

                # Name of artist
                xpath_to_name_artist = "/html/body/div[3]/div[3]/div/div/div[6]/div/header/div[2]/div[1]/div[1]/h2/span/a/text()"
                url_to_artist_name = "".join(read_html_to_artwork.xpath(xpath_to_name_artist))
                name_artist_cleaned = re.sub(r'\W+', '_', url_to_artist_name)
                print(name_artist_cleaned, name_artwork_cleaned)

                urllib.request.urlretrieve(url_to_image_cleaned, folder + "\\" + name_artist_cleaned + "_" + name_artwork_cleaned + ".jpg")

                time.sleep(SCROLL_PAUSE_TIME)


            # Scroll down to bottom
            images_website.execute_script("window.scrollTo(0, document.body.scrollHeight);")

            # Wait to load page
            time.sleep(SCROLL_PAUSE_TIME)

            # Calculate new scroll height and compare with last scroll height
            new_height_image_website = images_website.execute_script("return document.body.scrollHeight")
            if new_height_image_website == last_height_image_website:
                break
            last_height_image_website = new_height_image_website        
        
        
    
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
    
driver.close()
```
